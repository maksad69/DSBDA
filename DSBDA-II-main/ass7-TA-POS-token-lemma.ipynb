```python
text = """Tokenization is the first step in text analytics.
It involves breaking down text into meaningful units."""
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
sentence=sent_tokenize(text)
sentence
```

```python
words=word_tokenize(text)
text
```

```python
from nltk.corpus import stopwords
Stopwords=set(stopwords.words('english'))
Stopwords
```

```python
for sent in sentence:
  Word=word_tokenize(sent)
 # print(Word)
  filtered=[word.lower() for word in Word if word.lower()not in Stopwords]
  print(filtered)
  print(nltk.pos_tag(filtered))
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps=PorterStemmer()
lemmatizer=WordNetLemmatizer()

```

```python
for word in words :
  print(ps.stem(word  ))
```

```python
for word in words :
  print(f'{lemmatizer.lemmatize(word)}')
```
